```python
import cv2
from matplotlib import pyplot as plt
from skimage import feature
from skimage.filters import gaussian
import imutils
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
import cv2
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
image_path = '/content/3.jpeg'
image = cv2.imread(image_path)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray_image, scaleFactor=2, minNeighbors=5, minSize=(30, 30))
mask = np.zeros_like(image)
for (x, y, w, h) in faces:
    cv2.rectangle(mask, (x+70, y+10), (x+310, y+345), (255, 255, 255), thickness =cv2.FILLED)
result = cv2.bitwise_and(image, mask)
plt.imshow(result)
```

```python
gray = cv2.cvtColor(result, cv2.COLOR_BGR2HSV)
bfilter = cv2.bilateralFilter(gray, 11,190,170)
edged = cv2.Canny(bfilter, 800, 10)
plt.imshow(cv2.cvtColor(edged, cv2.COLOR_BGR2RGB))
```

```python
border_color = (0, 0, 0)  # Black color
border_thickness = 2  # Adjust the thickness of the border if needed
for (x, y, w, h) in faces:
    cv2.rectangle(edged, (x+70, y+10), (x+310, y+345), border_color, border_thickness)
```
